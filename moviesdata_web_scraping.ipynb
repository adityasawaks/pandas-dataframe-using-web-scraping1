```python
namee=[]
final=pd.DataFrame()
for j in range(1,89):
  webpage=requests.get('https://moviesverse.id/page/{}/'.format(j)).text
  soup=BeautifulSoup(webpage,'lxml')
  company=soup.find_all('h2')

  for i in company:
    namee.append(i.text)
```

```python
finaldf=pd.DataFrame({'name':namee})
```
